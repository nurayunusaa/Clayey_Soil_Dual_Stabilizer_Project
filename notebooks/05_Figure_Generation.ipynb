# 05_Figure_Generation
Generate PNG figures for manuscript

```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('../data/processed/fused_dataset.csv')

# Figure1: UCS vs CBR
plt.figure(figsize=(6,4))
sns.scatterplot(x='CBR', y='UCS', data=df)
plt.xlabel('CBR (%)')
plt.ylabel('UCS (kPa)')
plt.title('UCS vs CBR for Stabilized Soil')
plt.savefig('../main_paper/figures/Figure1_UCS_CBR.png', dpi=300)
plt.close()

# Placeholder: Add other figures similarly (Figures 2â€“7) using relevant columns and plots
```
