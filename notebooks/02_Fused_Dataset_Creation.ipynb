# 02_Fused_Dataset_Creation
Create fused dataset combining real and synthetic data

```
import pandas as pd

# Load datasets
real = pd.read_csv('../data/raw/UCS_CBR_data.csv')
synthetic = pd.read_csv('../data/raw/synthetic_dataset.csv')  # placeholder

# Combine datasets
fused = pd.concat([real, synthetic], ignore_index=True)

# Save fused dataset
fused.to_csv('../data/processed/fused_dataset.csv', index=False)
print('Fused dataset saved:', fused.shape)
```
